```python
# collecting the hashtags

def hashtag_extract(x):
    hashtags = []
    
    for i in x:
        ht = re.findall(r"#(\w+)", i)
        hashtags.append(ht)

    return hashtags
```

```python
# extracting hashtags from non racist/sexist tweets
HT_regular = hashtag_extract(result['text'][result['label'] == 0])

# extracting hashtags from racist/sexist tweets
HT_negative = hashtag_extract(result['text'][result['label'] == 1])

# unnesting list
HT_regular = sum(HT_regular,[])
HT_negative = sum(HT_negative,[])
```

```python
#removing twitter handels and stopwords
def remove_pattern(text,pattern):
    
    # re.findall() finds the pattern i.e @user and puts it in a list for further task
    r = re.findall(pattern,text)
    
    # re.sub() removes @user from the sentences in the dataset
    for i in r:
        text = re.sub(i,"",text)


        words = text.split(" ")
        words = [lemma.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
        text = ' '.join(words)
    
    return text
```

```python
result['Tidy_Tweets'] = np.vectorize(remove_pattern)(result['text'], "@[\w]*")
result.head()
```
